```python
from pandas.tools.plotting import scatter_matrix
scatter_matrix(recent_grads[["ShareWomen","Unemployment_rate"]])
```

```python
fig = plt.figure(figsize=(8,8))
ax1 = fig.add_subplot(2,2,1)
recent_grads['ShareWomen'].hist()
ax2 = fig.add_subplot(2,2,2)
ax2.scatter(recent_grads['Unemployment_rate'],recent_grads['ShareWomen'], color='darkblue', marker='+') 
ax3 = fig.add_subplot(2,2,3)
ax3.scatter(recent_grads['ShareWomen'], recent_grads['Unemployment_rate'], color='darkgreen', marker='+')
ax4 = fig.add_subplot(2,2,4)
recent_grads['Unemployment_rate'].hist()
plt.show()
```

```python
fig = plt.figure(figsize=(8,8))
ax1 = fig.add_subplot(2,2,1)
recent_grads['ShareWomen'].hist()
ax2 = fig.add_subplot(2,2,2)
ax2.scatter(recent_grads['Unemployment_rate'],recent_grads['ShareWomen'], color='darkblue', marker='+') 
ax3 = fig.add_subplot(2,2,3)
ax3.scatter(recent_grads['ShareWomen'], recent_grads['Unemployment_rate'], color='darkgreen', marker='+')
ax4 = fig.add_subplot(2,2,4)
recent_grads['Unemployment_rate'].hist()
plt.show()
```

```python
fig = plt.figure(figsize=(8,8))
fig.subplots_adjust(wspace = 0, hspace = 0)
ax1 = fig.add_subplot(2,2,1)
recent_grads['ShareWomen'].hist()
ax2 = fig.add_subplot(2,2,2)
ax2.scatter(recent_grads['Unemployment_rate'],recent_grads['ShareWomen'], color='darkblue', marker='+') 
ax3 = fig.add_subplot(2,2,3)
ax3.scatter(recent_grads['ShareWomen'], recent_grads['Unemployment_rate'], color='darkgreen', marker='+')
ax4 = fig.add_subplot(2,2,4)
recent_grads['Unemployment_rate'].hist()


ax1.xaxis.set_visible(False)
ax2.yaxis.set_visible(False)
ax2.xaxis.set_visible(False)
ax4.yaxis.set_visible(False)

ax1.get_yaxis().tick_left()
ax2.get_yaxis().tick_left()
ax3.get_yaxis().tick_left()
ax4.get_yaxis().tick_left()


ax1.set_xlabel("ShareWomen")
ax1.set_ylabel("ShareWomen")
ax3.set_xlabel("ShareWomen")
ax3.set_ylabel("Unemployment_rate")
ax4.set_ylabel("Unemployment_rate")
ax4.set_xlabel("Unemployment_rate")


ax1.set_ylim([0,30])
ax2.set_xlim([0.0,0.2])
ax3.set_xlim([0.0,1.0])
ax3.set_ylim([0.0,0.2])
ax4.set_xlim([0.0,0.2])

ax1.set_yticklabels([0, 5, 10, 15, 20, 25, 30])
ax3.set_yticklabels([0.00, 0.05, 0.10, 0.15])
ax3.set_xticklabels([0.0, 0.2, 0.4, 0.6, 0.8], rotation=90)
ax4.set_xticklabels([0.00, 0.05, 0.10, 0.15, 0.20], rotation=90)
plt.show()
```

```python
recent_grads['ShareMen'] = recent_grads['Men']/recent_grads['Total']
arts = recent_grads[recent_grads["Major_category"] == "Arts"]

fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1,1,1)

import numpy as np

locs = np.arange(0,len(arts))
print(locs)
width = 0.35
bar_1 = plt.bar(left = locs,height = arts['ShareMen'],width = width,data = recent_grads)
ax.set_xticklabels(arts['Major'].tolist(),rotation= 90)
offset_locs = locs+0.35
bar_2 =plt.bar(left = offset_locs,height = arts['ShareWomen'],width = width,data = recent_grads,color = 'green')
ax.set_xticks(offset_locs)
plt.legend((bar_1,bar_2),("ShareMen","ShareWomen"),loc= "upper left")
plt.grid()
```
